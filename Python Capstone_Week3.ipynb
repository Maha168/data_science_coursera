##### 1.Getting the Dataframe from Wiki

```python
import pandas as pd
import requests 
#Getting the Dataframe from web page
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',header=0)
print(df[0])
df1=df[0]
df1
```

```python
#Removing Not assigned Borough
df1=df1[df1.Borough != 'Not assigned']
df1
```

##### 2.Grouping and appending the values in Neighbourhood column


```python
#Grouping and appending the values in Neighbourhood column
df2=pd.DataFrame(df1.groupby('Postcode')['Neighbourhood'].apply(lambda x: "%s" % ', '.join(x)))

#Merging the dataframes and modifying it to get the single df to work on 

df3=pd.merge(df2,df1,on='Postcode',how='right')
df3.drop(columns='Neighbourhood_y',axis=0,inplace=True)
df3.drop_duplicates(inplace=True)
df3= df3[['Postcode','Borough','Neighbourhood_x']]
df3.rename(columns={'Neighbourhood_x':'Neighbourhood'},inplace=True)
df3
```

##### 3.Getting the not assigned value in "Neighbourhood" and replacing it with corresponding "Borough" value


```python
#Getting the not assigned Neighbourhood and replacing it with Borough value
df3['Neighbourhood'].replace("Not assigned",df3.Borough,inplace=True)
df3
```
